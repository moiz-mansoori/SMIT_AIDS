```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Overall Rating'], bins=10, kde=True, color='blue') # Plot the distribution of overall ratings
plt.title('Distribution of Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show() 
```

## Average Ratings by Travel Class

```python
avg_ratings_by_class = df.groupby('Travel Class')['Overall Rating'].mean().sort_values() # Calculate average ratings by travel class and sort them
print(avg_ratings_by_class) # Display the average ratings by travel class
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_ratings_by_class.index, y=avg_ratings_by_class.values, palette='viridis')
plt.title('Average Ratings by Travel Class')
plt.xlabel('Travel Class')
plt.ylabel('Average Rating')
plt.show()
```

## Correlation Heatmapplt


```python
plt.figure(figsize=(10, 6))

# Filter numeric columns only
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Sentiment Analysis on Reviews

```python
# !pip install textblob # Install TextBlob for sentiment analysis
from textblob import TextBlob  # Import TextBlob for sentiment analysis

# Define a function to calculate sentiment polarity
def get_sentiment(review):
    return TextBlob(str(review)).sentiment.polarity

# Apply the function to the 'Review' column
df['Sentiment'] = df['Review'].apply(get_sentiment)

# Plot sentiment distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Sentiment'], bins=20, kde=True, color='green')
plt.title('Sentiment Polarity Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

## Ratings vs Sentiment


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Sentiment', y='Overall Rating', data=df, alpha=0.6) # Scatter plot of sentiment polarity vs overall rating
plt.title('Sentiment Polarity vs Overall Rating')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Overall Rating')
plt.show()
```

## Advanced Statistical Analysis

```python
from scipy import stats

# Filter rows with either 'Business Class' or 'Economy Class' and no NaNs in Overall Rating
filtered_df = df[df['Travel Class'].isin(['Business Class', 'Economy Class'])]
filtered_df = filtered_df[['Travel Class', 'Overall Rating']].dropna()

# Separate the groups
business = filtered_df[filtered_df['Travel Class'] == 'Business Class']['Overall Rating']
economy = filtered_df[filtered_df['Travel Class'] == 'Economy Class']['Overall Rating']

# Perform independent t-test
t_stat, p_val = stats.ttest_ind(business, economy, nan_policy='omit')

print(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
```
